# Q2:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://stockanalysis.com/stocks/tsla/revenue/'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')  # Adjust if necessary to locate the correct table

headers = [header.text for header in table.find_all('th')]
data = []

for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    data.append([column.text.strip() for column in columns])

tesla_revenue = pd.DataFrame(data, columns=headers)
tesla_revenue.tail()
```

# Q 5:

```python
import plotly.graph_objects as go

# Fetch Tesla stock data using yfinance
tesla_data = yf.download('TSLA', start='2022-01-01', end='2023-01-01')

# Define the make_graph function
def make_graph(data, title):
    fig = go.Figure()

    # Add a trace for the closing price
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Add titles and labels
    fig.update_layout(title=title,
                      xaxis_title='Date',
                      yaxis_title='Price (USD)',
                      template='plotly_dark')

    # Show the figure
    fig.show()

# Use the function to plot Tesla stock data
make_graph(tesla_data, title='Tesla Stock Price (2022)')
```

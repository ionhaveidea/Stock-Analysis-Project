```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL for Tesla's revenue on StockAnalysis
url = 'https://stockanalysis.com/stocks/tsla/revenue/'

# Fetch the content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the revenue table
table = soup.find('table')  # Adjust if necessary to locate the correct table

# Extract the header
headers = [header.text for header in table.find_all('th')]
data = []

# Extract the rows
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    data.append([column.text.strip() for column in columns])

# Create DataFrame
tesla_revenue = pd.DataFrame(data, columns=headers)

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL for GameStop's revenue on Companies Market Cap
url = 'https://companiesmarketcap.com/gamestop/revenue/'

# Fetch the content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the revenue table
table = soup.find('table')

# Check if the table is found
if table is not None:
    # Extract the header
    headers = [header.text for header in table.find_all('th')]
    data = []

    # Extract the rows
    for row in table.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        data.append([column.text.strip() for column in columns])

    # Create DataFrame
    gme_revenue = pd.DataFrame(data, columns=headers)

    # Display the last five rows
    print(gme_revenue.tail())
else:
    print("Table not found. Please check the URL or the page structure.")
```

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go

# Fetch Tesla stock data using yfinance
tesla_data = yf.download('TSLA', start='2022-01-01', end='2023-01-01')

# Define the make_graph function
def make_graph(data, title):
    fig = go.Figure()

    # Add a trace for the closing price
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Add titles and labels
    fig.update_layout(title=title,
                      xaxis_title='Date',
                      yaxis_title='Price (USD)',
                      template='plotly_dark')

    # Show the figure
    fig.show()

# Use the function to plot Tesla stock data
make_graph(tesla_data, title='Tesla Stock Price (2022)')
```
